```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(Path(r"C:\Users\nasir\OneDrive\Desktop\Fintech\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources/sfo_neighborhoods_census_data.csv")).drop_duplicates()

# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

```python
# Create a numerical aggregation that groups the data by the year and then averages the results.
#housing_units_by_year = sfo_data_df.groupby("year").mean()

housing_units_by_year = (sfo_data_df[["year", "housing_units"]]
    .groupby("year")
    .mean()
    .sort_values("housing_units")
)
# View the result
housing_units_by_year
```

```python
# Create an interactive bar plot with hvplot where the x-axis is the "Provider State" 
# and the y-axis is the "Average Total Payments"
housing_units_by_year.hvplot .bar(
    x="year", 
    y="housing_units", 
    title="housing_units in Sanfrancisco - 2010 to 2016" , 
).opts(yformatter='%.0f')
```

```python
# Filter out the housing_units column, creating a new DataFrame 
# Keep only sale_price_sqr_foot and gross_rent averages per year

prices_square_foot_by_year = (sfo_data_df[["year", "sale_price_sqr_foot"]]
    .groupby("year")
    .mean()
    .sort_values("year")
)

# Review the resulting DataFrame
prices_square_foot_by_year
```

```python
# # Plot prices_square_foot_by_year. 

prices_square_foot=prices_square_foot_by_year.hvplot(
    x='year', 
    y='sale_price_sqr_foot',
    title="sale_price_sqr_foot" ,
)
prices_square_foot
```

```python
 # Filter out the gross rent column, creating a new DataFrame 
# Keep only  gross_rent averages per year

gross_rent_by_year= ( sfo_data_df[["year", "gross_rent"]]
    .groupby("year")
    .mean()
    .sort_values("year")
)

# Review the resulting DataFrame
gross_rent_by_year
```

```python
# Plot average gross_rent_by_year.
gross_rent=gross_rent_by_year.hvplot(
    x='year', 
    y='gross_rent',
    xformatter='%.0f',
    hover_color="orange",
    title="gross_rent" ,
)
gross_rent
```
