```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(Path(r"C:\Users\nasir\OneDrive\Desktop\Fintech\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources/sfo_neighborhoods_census_data.csv")).drop_duplicates()

# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

```python
# Create a numerical aggregation that groups the data by the year and then averages the results.
#housing_units_by_year = sfo_data_df.groupby("year").mean()

housing_units_by_year = (sfo_data_df[["year", "housing_units"]]
    .groupby("year")
    .mean()
    .sort_values("housing_units")
)
# View the result
housing_units_by_year
```

```python
# Create an interactive bar plot with hvplot where the x-axis is the "Provider State" 
# and the y-axis is the "Average Total Payments"
housing_units_by_year.hvplot .bar(
    x="year", 
    y="housing_units", 
    title="housing_units in Sanfrancisco - 2010 to 2016" , 
).opts(yformatter='%.0f')
```

```python
# Filter out the housing_units column, creating a new DataFrame 
# Keep only sale_price_sqr_foot and gross_rent averages per year

prices_square_foot_by_year = (sfo_data_df[["year", "sale_price_sqr_foot"]]
    .groupby("year")
    .mean()
    .sort_values("year")
)

# Review the resulting DataFrame
prices_square_foot_by_year
```

```python
 # Filter out the gross rent column, creating a new DataFrame 
# Keep only  gross_rent averages per year

gross_rent_by_year= ( sfo_data_df[["year", "gross_rent"]]
    .groupby("year")
    .mean()
    .sort_values("year")
)

# Review the resulting DataFrame
gross_rent_by_year
```

```python
# Plot average gross_rent_by_year.
gross_rent=gross_rent_by_year.hvplot(
    x='year', 
    y='gross_rent',
    xformatter='%.0f',
    hover_color="orange",
    title="gross_rent" ,
)
gross_rent
```

```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
neighborhoods_coordinates_df = pd.read_csv(Path(r"C:\Users\nasir\OneDrive\Desktop\Fintech\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources/neighborhoods_coordinates.csv")).drop_duplicates()
sfo_data_df = pd.read_csv(Path(r"C:\Users\nasir\OneDrive\Desktop\Fintech\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources/sfo_neighborhoods_census_data.csv")).drop_duplicates()
housing_per_year_df = pd.read_csv(Path(r"C:\Users\nasir\OneDrive\Desktop\Fintech\UTOR-VIRT-FIN-PT-08-2023-U-LOLC\06-PyViz\Unit 6 Homework\Instructions\Starter_Code\Resources/housing_per_year.csv")).drop_duplicates()

# Review a sample of ten rows of the DataFrame
display(sfo_data_df.sample(10))
display(housing_per_year_df.head())
```

```python
# Create a DataFrame consisting of the CityName, PopulationCount, Latitude and Longitude columns
neighborhood_sale_price_sqr_foot_df = sfo_data_df[["neighborhood", "sale_price_sqr_foot"]]

# Review the DataFrame
neighborhood_sale_price_sqr_foot_df.head()
```

```python
# Create a numerical aggregation that groups the data by the year and then averages the results.
#housing_units_by_year = sfo_data_df.groupby("year").mean()

neighborhood_sale_price_Ave_df = (neighborhood_sale_price_sqr_foot_df
    .groupby("neighborhood")
    .mean()
    .sort_values("neighborhood")
)
# View the result
neighborhood_sale_price_Ave_df
```

```python
# Review the DataFrame
neighborhoods_coordinates_df
```

```python
# Set the index as the 'Neighborhood' column
neighborhoods_coordinates_df = neighborhoods_coordinates_df.set_index(neighborhoods_coordinates_df["Neighborhood"])

neighborhoods_coordinates_df.head()
```

```python
# Concatenate the two DataFrames by columns and perform an inner join
all_neighborhoods_df = pd.concat([neighborhoods_coordinates_df, neighborhood_sale_price_Ave_df], axis="columns", join="inner")
all_neighborhoods_df
```

```python
# to reset index and review data
all_neighborhoods_df=all_neighborhoods_df.reset_index(drop=True)
all_neighborhoods_df
```

```python
# Create a plot to analyze neighborhood info
Average_sale_price_by_neighborhood = all_neighborhoods_df.hvplot.points(
    'Lon', 
    'Lat', 
    geo=True, 
    color='Neighborhood',
    size = 'sale_price_sqr_foot',
    alpha=0.8,
    tiles='OSM',
    frame_width = 700,
    frame_height = 500
    )
# Show the plot
Average_sale_price_by_neighborhood
```
